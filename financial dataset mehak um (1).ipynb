```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "C:\\Users\\MEHAK KAUR\\Downloads\\Financial Analytics data.csv"
df = pd.read_csv(file_path)

# Step 2: Data Preprocessing
# Check for missing values
if df.isnull().any().any():
    print("Warning: Missing values detected!")
    # Handle missing values if needed

# Convert data types if necessary
# df['Mar Cap  Crore'] = df['Mar Cap  Crore'].astype(float)
# df['Sales Qtr  Crore'] = df['Sales Qtr  Crore'].astype(float)

# Step 3: Calculate key metrics
# Market Capitalization to Sales Ratio
df['MC_to_Sales_Ratio'] = df['Mar Cap - Crore'] / df['Sales Qtr - Crore']

# Calculate Yearly Sales assuming quarterly sales are approximately equal across quarters
df['Yearly_Sales'] = df['Sales Qtr - Crore'] * 4

# Calculate Average Market Capitalization
avg_mc = df['Mar Cap - Crore'].mean()

# Calculate Average Sales
avg_sales = df['Sales Qtr - Crore'].mean()

# Step 4: Explore relationships between attributes
# Correlation between Market Capitalization and Sales
correlation = df[['Mar Cap - Crore', 'Sales Qtr - Crore']].corr().iloc[0, 1]

# Step 5: Visualize the findings

# Scatter plot: Market Capitalization vs Sales
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Sales Qtr - Crore', y='Mar Cap - Crore', data=df)
plt.title('Market Capitalization vs Quarterly Sales')
plt.xlabel('Quarterly Sales (Crore)')
plt.ylabel('Market Capitalization (Crore)')
plt.axhline(avg_mc, color='r', linestyle='--', label=f'Average MC: {avg_mc:.2f} Crore')
plt.axvline(avg_sales, color='g', linestyle='--', label=f'Average Sales: {avg_sales:.2f} Crore')
plt.legend()
plt.show()

# Bar plot: Market Capitalization to Sales Ratio by Company
plt.figure(figsize=(15, 8))
sns.barplot(x='Name', y='MC_to_Sales_Ratio', data=df.sort_values(by='MC_to_Sales_Ratio', ascending=False))
plt.title('Market Capitalization to Sales Ratio by Company')
plt.xlabel('Company')
plt.ylabel('MC to Sales Ratio')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Print key metrics and findings
print(f"Correlation between Market Capitalization and Quarterly Sales: {correlation}")
print(f"\nAverage Market Capitalization: {avg_mc:.2f} Crore")
print(f"Average Quarterly Sales: {avg_sales:.2f} Crore")
print("\nKey Metrics:")
print(df[['Name', 'MC_to_Sales_Ratio', 'Yearly_Sales']].sort_values(by='MC_to_Sales_Ratio', ascending=False))

# Additional Insights
# Companies with MC to Sales Ratio greater than 50
high_ratio_companies = df[df['MC_to_Sales_Ratio'] > 50]
print("\nCompanies with MC to Sales Ratio > 50:")
print(high_ratio_companies[['Name', 'MC_to_Sales_Ratio', 'Yearly_Sales']])

```
